```python
import numpy as np
import matplotlib.pyplot as plt
from drawdata import ScatterWidget
import matplotlib.pyplot as plt
```

```python
df = widget.data_as_pandas

df = df.drop("color", axis=1)

df = df.to_numpy()

for i in range(df.shape[0]):
  if df[i][2] == "a":
    df[i][2] = 0
  else:
    df[i][2] = 1
print(df)


np.random.shuffle(df)
split = int(df.shape[0] * 0.9)
X_train = df[:split, :2].T
X_train = np.vstack((X_train, X_train[0, :]**2, X_train[1, :]**2 ))
Y_train = df[:split, 2].T
X_val = df[split:, :2].T
Y_val = df[split:, 2].T
```

```python
def sigmoid(x):
  x = x.astype("float")
  return 1 / (1 + np.exp(-x))
```

```python
def derivative_of_tanh(x):
  x = x.astype("float")
  return 1 - np.tanh(x)**2
```

```python
def init(X, layers=[3, 3, 1]):
  W = {}
  alpha = {}
  beta = {}
  W[0] = np.random.randn(layers[0], X.shape[0])
  alpha[0] = np.ones((layers[0], 1), dtype="float")
  beta[0] = np.ones((layers[0], 1), dtype="float")
  for i in range(1, len(layers)):
    if i != len(layers) - 1:
      W[i] = np.random.randn(layers[i], W[i-1].shape[0])
      alpha[i] = np.ones((layers[i], 1), dtype="float")
      beta[i] = np.ones((layers[i], 1), dtype="float")
    else:
      W[i] = np.random.randn(layers[i], W[i-1].shape[0])
  return W, alpha, beta
```

```python
def batch_norm(x):
  mean = np.mean(x, axis=1, keepdims=True, dtype="float")
  var = np.var(x, axis=1, keepdims=True, dtype="float")
  norm = (x-mean)/np.sqrt(var + 1e-8)
  return norm, mean, var
```

```python
def cost(A, Y):
  return -np.sum(Y * np.log(A) + (1.00 - Y) * np.log(1.00 - A)) / A.shape[1]
```

```python
def forward(X, W, alpha, beta):
  Z, T, K, A = {}, {}, {}, {}
  mean, var = {}, {}
  Z[0] = np.dot(W[0], X)
  T[0], mean[0], var[0] = batch_norm(Z[0])
  K[0] = T[0] * alpha[0] + beta[0]
  A[0] = np.tanh(K[0].astype("float"))
  for i in range(1, len(W)):
    if i!= (len(W)-1):
      Z[i] = np.dot(W[i], A[i-1])
      T[i], mean[i], var[i] = batch_norm(Z[i])
      K[i] = T[i] * alpha[i] + beta[i]
      A[i] = np.tanh(K[i].astype("float"))
    else:
      Z[i] = np.dot(W[i], A[i-1])
      A[i] = sigmoid(Z[i])
  return Z, T, K, A, mean, var
```

```python
def backward(X, Y, W, alpha, Z, T, K, A, mean, var ):
  l = len(W)-1
  m = X.shape[1]
  dW, dalpha, dbeta = {}, {}, {}
  dZ = A[l] - Y
  dW[l] = np.dot(dZ, A[l-1].T) / m
  dA = np.dot(W[l].T, dZ)

  for i in range(l-1, 0, -1):
    dK = derivative_of_tanh(K[i]) * dA

    dalpha[i] = np.sum(dK * T[i], axis=1, keepdims=True)
    dbeta[i] = np.sum(dK, axis=1, keepdims=True)

    dT = alpha[i] * dK

    dsigma = np.sum(dT * (Z[i] - mean[i]) * ((-1/2) * (var[i] + 1e-8)**(-3/2)) , axis = 1, keepdims=True)
    dmean  = np.sum(dT * (-1 / (np.sqrt(var[i] + 1e-8)) ), axis=1, keepdims=True) + dsigma * (np.sum(-2 * (Z[i] - mean[i]), axis = 1, keepdims=True)/m)

    dZ = dT * (1/np.sqrt(var[i] + 1e-8)) + dsigma * (((2 * (Z[i] - mean[i])) /m)) + dmean / m

    dW[i] = np.dot(dZ, A[i-1].T) / m
    dA = np.dot(W[i].T, dZ)



  dK = derivative_of_tanh(K[0]) * dA

  dalpha[0] = np.sum(dK * T[0], axis=1, keepdims=True)
  dbeta[0] = np.sum(dK, axis=1, keepdims=True)

  dT = alpha[0] * dK

  dsigma = np.sum(dT * (Z[0] - mean[0]) * (-1/2) * (var[0] + 1e-8)**(-3/2) , axis = 1, keepdims=True)
  dmean  = np.sum(dT * (-1 / (np.sqrt(var[0] + 1e-8)) ), axis=1, keepdims=True) + dsigma * (np.sum(-2 * (Z[0] - mean[0]), axis = 1, keepdims=True)/m)

  dZ = dT * (1/np.sqrt(var[0] + 1e-8)) + dsigma * (((2 * (Z[0] - mean[0])) /m)) + dmean / m

  dW[0] = np.dot(dZ, X.T) / m

  return dW, dalpha, dbeta


```

```python
def update(W, alpha, beta, dW, dalpha, dbeta, learning_rate=0.01):
  for i in range(len(W)):
    if i != len(W) - 1:
      W[i] = W[i] - learning_rate * dW[i]
      alpha[i] = alpha[i] - learning_rate * dalpha[i]
      beta[i] = beta[i] - learning_rate * dbeta[i]
    else:
      W[i] = W[i] - learning_rate * dW[i]
  return W, alpha, beta
```

```python
W, alpha, beta = init(X_train, layers=[ 10 , 10, 10, 10, 3,  1])
Z, T, K, A, mean, var = forward(X_train, W, alpha, beta)
```

```python
print(cost(A[len(W)-1], Y_train))
```

```python
c = []
for i in range(10000):
  Z, T, K, A, mean, var = forward(X_train, W, alpha, beta)
  dW, dalpha, dbeta = backward(X_train, Y_train, W, alpha, Z, T, K, A, mean, var)
  W, alpha, beta = update(W, alpha, beta, dW, dalpha, dbeta, learning_rate=0.06)
  c.append(cost(A[len(W)-1], Y_train))
print(cost(A[len(W)-1], Y_train))
```

```python
plt.figure(figsize=(12, 6))
plt.plot(range(10000), c, linestyle='-', color='b', label="Temperature (°C)")

# Customizing the plot
plt.xlabel("Iteration")
plt.ylabel("Cost")
```
